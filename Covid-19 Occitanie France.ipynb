# Covid-19 Occitanie France

Initial Data

```python
import pandas as pd
import plotly
import plotly.offline as py

from typing import List, Optional

from plotly import graph_objs as go
from plotly import figure_factory as ff
from plotly import tools
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

plotly.offline.init_notebook_mode(connected=True)
```

```python
df = pd.read_csv('./Covid19OccitanieFrance.csv')
df
```

```python
positifs_actuellement = []
increase_positifs_actuellement = []
percentage_positifs_actuellement = []

hospitalises_en_cours = [int(h) for h in df[["Hospitalization en cours"]].values]
increase_hospitalises = []

positifs_non_hospitalises = []
increase_positifs_non_hospitalises = []

guerie_v = [int(g) for g in df[["Guerie"]].values]
increase_guerie = []

deces_v = [int(d) for d in df[["Deces"]].values]
increase_deces = []

c = 0
for index, row in df[["Positifs total", "Guerie", "Deces", "Hospitalization en cours"]].iterrows():
    positfs_total = row[0]
    guerie = row[1]
    deces = row[2]
    hospitalises = row[3]
    positifs_actuellement_ = positfs_total - guerie - deces
    positifs_actuellement.append(positifs_actuellement_)
    
    positifs_non_hospitalises.append(positifs_actuellement_ - hospitalises)

    if c == 0:
        increase_positifs_actuellement.append(positifs_actuellement[c])
        
        increase_positifs_non_hospitalises.append(increase_positifs_actuellement[c])
        
        increase_hospitalises.append(hospitalises_en_cours[c])
        increase_guerie.append(guerie_v[c])
        increase_deces.append(deces_v[c])
    else:
        increase_positifs_actuellement.append(positifs_actuellement[c] - positifs_actuellement[c - 1])
        
        increase_positifs_non_hospitalises.append(
            increase_positifs_actuellement[c] - increase_positifs_actuellement[c - 1])
        
        increase_hospitalises.append(hospitalises_en_cours[c] - hospitalises_en_cours[c - 1])
        increase_guerie.append(guerie_v[c] - guerie_v[c - 1])
        increase_deces.append(deces_v[c] - deces_v[c - 1])
        
    percentage_positifs_actuellement.append(increase_positifs_actuellement[c]/positifs_actuellement[c]*100)
    
    c += 1

df["Positifs actuellement"] = positifs_actuellement
df["Increase Positifs actuellement"] = increase_positifs_actuellement
df["Percentage Increase Positifs actuellement"] = percentage_positifs_actuellement

df["Increase Guerie"] = increase_guerie

df["Increase Deces"] = increase_deces

df["Increase hospitalisés"] = increase_hospitalises

df["Positifs non hospitalisés"] = positifs_non_hospitalises
df["Increase Positifs non hospitalisés"] = increase_positifs_non_hospitalises

df
```

```python
def create_2d_plot(plot_df: pd.DataFrame, x_name: str, y_name: str, fig: Optional[go.Figure] = None):
    """Create 2D plot."""
    trace = go.Scatter(
        x=plot_df[x_name],
        y=plot_df[y_name],
        mode="lines+markers",
        marker=dict(
            size=6,
            color=[float(n) for n in plot_df[y_name].values],  # set color to an array/list of desired values
            opacity=0.8,
        ),
        name=y_name,
    )
    
    annotations = []
    c = 0

    for (x, y) in zip(plot_df[x_name], plot_df[y_name]):
        annotations.append(
            dict(
                showarrow=False,
                x=x,
                y=y,
                text=str(plot_df[y_name].values[c]),
                xanchor="left",
                xshift=15,
                opacity=0.7,
            )
        )
        c += 1
   
    return trace, annotations
```

```python
trace1, annotations1 = create_2d_plot(plot_df=df, x_name="Date", y_name="Positifs total")
trace2, annotations2 = create_2d_plot(plot_df=df, x_name="Date", y_name="Positifs actuellement")

data = [trace1, trace2]

annotations = [annotations1[len(annotations1) - 1]] + [annotations2[len(annotations2) - 1]]

layout = go.Layout(
    title=f"Covid19-France-Occitanie Positifs total vs Positifs actuels",
    xaxis=dict(title="Time"),
    legend=dict(orientation="h", y = -0.3, yanchor="top"),
    annotations=annotations,
    showlegend=True,
) 
fig = go.Figure(data=data, layout=layout)
    
iplot(fig, filename="covid19-Occitanie")
```
